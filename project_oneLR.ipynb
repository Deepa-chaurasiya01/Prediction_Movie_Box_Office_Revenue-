```python
import pandas as pd
from pandas import DataFrame as df
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python

data = pd.read_csv('/content/cost_revenue_clean (1).csv')
data.describe()
# .describe() :computes and displays summary statistics for a Python dataframe.
```

```python
5.03e3
```

```python
X = df(data, columns=['production_budget_usd'])
y = df(data, columns=['worldwide_gross_usd'])
```

```python
plt.figure(figsize = (10,6))
plt.scatter(X,y,alpha = 0.3) #alpha for the transparency
plt.title("Flim Cost Vs Global Revenue")
plt.xlabel("production_budget_usd")
plt.ylabel("worldwide_gross_usd")
plt.ylim(0,3000000000)
plt.xlim(0,450000000)
plt.show()
```

```python
reg = LinearRegression()
reg.fit(X,y)
```

Slpoe coefficient.

```python
reg.coef_ # theta_1
```

we have 3.11 which means we have positive realtionship between bugdeget and revenue and also for each dollar that we spend on producing the moive, we should get around 3.1 dollars in revenue in return.

```python
# intercept
reg.intercept_
```

well this intercept is litrally telling us is that a movie with a budget of zero would actually lose over 7 million dollars

```python
plt.figure(figsize = (10,6))
plt.scatter(X,y,alpha = 0.3) #alpha for the transparency
plt.plot(X, reg.predict(X), color='red', linewidth = 3)
plt.title("Flim Cost Vs Global Revenue")
plt.xlabel("production_budget_usd")
plt.ylabel("worldwide_gross_usd")
plt.ylim(0,3000000000)
plt.xlim(0,450000000)
plt.show()
```
